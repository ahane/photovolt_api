```python
def group_by_and_sum(df, timeAggr, localeAgg):
    cols = ['power_in_kW', timeAggr, localeAgg] 
    group_cols = [localeAgg, timeAggr] #this order is very important
    grouped_df = df[cols].groupby(group_cols).sum()
    grouped_df.reset_index(inplace = True)
    grouped_df.sort([timeAggr], ascending=True)
    return grouped_df
```

```python
group_by_and_sum(df, 'year-month', 'state')
```
